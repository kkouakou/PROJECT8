````python
# Configuration Windows
import os
os.environ['JAVA_HOME'] = r'C:\Program Files\Java\jdk-1.8'

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, avg, sum, count

# Créer la session Spark
spark = SparkSession.builder \
    .appName("TableauxPySparkPur") \
    .master("local[*]") \
    .getOrCreate()

# ============================================
# TABLEAU 1 : Données employés - Simple
# ============================================
print("╔══════════════════════════════════════════════════════════════════╗")
print("║                    TABLEAU DES EMPLOYÉS                          ║")
print("╠══════════════╤═══════╤══════════════╤════════════╤══════════════╣")
print("║ ID           │ Nom   │ Département  │ Âge        │ Salaire (€)  ║")
print("╠══════════════╪═══════╪══════════════╪════════════╪══════════════╣")

# Données
employes_data = [
    (1, "Alice Martin", "IT", 34, 55000),
    (2, "Bob Dupont", "RH", 45, 42000),
    (3, "Catherine Leroy", "Marketing", 29, 48000),
    (4, "David Bernard", "IT", 38, 52000),
    (5, "Élise Petit", "Direction", 41, 85000)
]

employes_df = spark.createDataFrame(
    employes_data, 
    ["ID", "Nom", "Département", "Âge", "Salaire"]
)

# Afficher ligne par ligne
rows = employes_df.collect()
for i, row in enumerate(rows):
    print(f"║ {row['ID']:<12} │ {row['Nom'][:10]:<10} │ {row['Département']:<12} │ {row['Âge']:<10} │ {row['Salaire']:<12,} ║")
    if i < len(rows) - 1:
        print("╟──────────────┼───────────┼──────────────┼────────────┼──────────────╢")

print("╚══════════════╧═══════════╧══════════════╧════════════╧══════════════╝")

# ============================================
# TABLEAU 2 : Avec statistiques
# ============================================
print("\n" + "="*80)
print("STATISTIQUES PAR DÉPARTEMENT")
print("="*80)

# Calcul des statistiques avec PySpark
stats_df = employes_df.groupBy("Département") \
    .agg(
        count("*").alias("Effectif"),
        avg("Âge").alias("Âge moyen"),
        avg("Salaire").alias("Salaire moyen"),
        sum("Salaire").alias("Masse salariale")
    ) \
    .withColumn("Âge moyen", col("Âge moyen").cast("int")) \
    .withColumn("Salaire moyen", col("Salaire moyen").cast("int")) \
    .orderBy("Salaire moyen", ascending=False)

# Afficher le tableau
print("┌──────────────┬────────────┬──────────────┬────────────────┬──────────────────┐")
print("│ Département  │ Effectif   │ Âge moyen    │ Salaire moyen  │ Masse salariale  │")
print("├──────────────┼────────────┼──────────────┼────────────────┼──────────────────┤")

stats_rows = stats_df.collect()
for i, row in enumerate(stats_rows):
    dept = row["Département"].ljust(12)
    eff = str(row["Effectif"]).ljust(10)
    age = str(row["Âge moyen"]).ljust(12)
    sal = f"{row['Salaire moyen']:,}".ljust(14)
    masse = f"{row['Masse salariale']:,}".ljust(16)
    print(f"│ {dept} │ {eff} │ {age} │ {sal} │ {masse} € │")
    if i < len(stats_rows) - 1:
        print("├──────────────┼────────────┼──────────────┼────────────────┼──────────────────┤")

print("└──────────────┴────────────┴──────────────┴────────────────┴──────────────────┘")

# ============================================
# TABLEAU 3 : Avec catégories
# ============================================
print("\n" + "╔══════════════════════════════════════════════════════════════════════════╗")
print("║                     CLASSIFICATION DES EMPLOYÉS                               ║")
print("╠══════╤══════════════════╤══════════════╤════════════╤══════════════╤══════════╣")
print("║ ID   │ Nom              │ Département  │ Âge        │ Salaire (€)  │ Catégorie║")
print("╠══════╪══════════════════╪══════════════╪════════════╪══════════════╪══════════╣")

# Ajouter une colonne catégorie
employes_cat = employes_df.withColumn(
    "Catégorie",
    when(col("Salaire") >= 80000, "★★★ Haut")
    .when(col("Salaire") >= 50000, "★★ Moyen+")
    .when(col("Salaire") >= 45000, "★ Moyen")
    .otherwise("Standard")
).orderBy(col("Salaire").desc())

cat_rows = employes_cat.collect()
for i, row in enumerate(cat_rows):
    print(f"║ {row['ID']:<4} │ {row['Nom'][:15]:<15} │ {row['Département']:<12} │ {row['Âge']:<10} │ {row['Salaire']:<10,} │ {row['Catégorie']:<8} ║")
    if i < len(cat_rows) - 1:
        print("╟──────┼──────────────────┼──────────────┼────────────┼──────────────┼──────────╢")

print("╚══════╧══════════════════╧══════════════╧════════════╧══════════════╧══════════╝")

# ============================================
# TABLEAU 4 : Résumé global
# ============================================
print("\n" + "*" * 90)

# Calculs globaux
total_employes = employes_df.count()
total_salaire = employes_df.agg(sum("Salaire")).first()[0]
moyenne_salaire = employes_df.agg(avg("Salaire")).first()[0]
moyenne_age = employes_df.agg(avg("Âge")).first()[0]

print("""
╔══════════════════════════════════════════════════════════════════════════════════╗
║                              RÉCAPITULATIF GLOBAL                                ║
╠══════════════════════════════════════════════╤═══════════════════════════════════╣
║ Nombre total d'employés                      │ {total:>30}                    ║
║ Masse salariale totale                       │ {total_sal:>30,.0f} €           ║
║ Salaire moyen                                │ {moy_sal:>30,.0f} €           ║
║ Âge moyen                                    │ {moy_age:>30.1f} ans           ║
║ Salaire minimum                              │ {min_sal:>30,.0f} €           ║
║ Salaire maximum                              │ {max_sal:>30,.0f} €           ║
╚══════════════════════════════════════════════╧═══════════════════════════════════╝
""".format(
    total=total_employes,
    total_sal=total_salaire,
    moy_sal=moyenne_salaire,
    moy_age=moyenne_age,
    min_sal=employes_df.agg({"Salaire": "min"}).first()[0],
    max_sal=employes_df.agg({"Salaire": "max"}).first()[0]
))

# ============================================
# TABLEAU 5 : Vue détaillée avec show() amélioré
# ============================================
print("\n" + "-" * 100)
print("VUE DÉTAILLÉE DU DATAFRAME SPARK")
print("-" * 100)

# Afficher le DataFrame avec show() standard
employes_df.show(truncate=False)

print("\n" + "=" * 100)
print("SESSION SPARK TERMINÉE AVEC SUCCÈS!")
print("=" * 100)

# Arrêter Spark
spark.stop()
````

````python
# ÉTAPE 3.5 - Charger les images (version corrigée)

import os
from pyspark.sql import Row

# Définir le chemin des images (corrigé)
TEST_DIR = "/home/sagemaker-user/fruits/fruits-360_dataset/fruits-360/Test"

# Lister les images
image_paths = []
for root, dirs, files in os.walk(TEST_DIR):
    for file in files:
        if file.endswith('.jpg'):
            image_paths.append(os.path.join(root, file))

print(f"📊 {len(image_paths)} images trouvées")

# Créer un RDD
rdd = sc.parallelize(image_paths[:10])  # 10 images pour test

# Lire les images
def read_image(path):
    with open(path, 'rb') as f:
        content = f.read()
    return Row(path=path, content=bytearray(content))

image_rows = rdd.map(read_image)
images_df = spark.createDataFrame(image_rows)

# Extraire les labels
from pyspark.sql.functions import split, element_at
images_with_labels = images_df.withColumn(
    "label", 
    element_at(split(images_df["path"], "/"), -2)
)

print("✅ Images chargées")
images_with_labels.show(5, truncate=False)
````

````python
# ÉTAPE 4 - Préparation des fonctions de prétraitement

import numpy as np
from PIL import Image
import io

print("🔧 Préparation des fonctions de prétraitement...")

# 1. Fonction de prétraitement d'une image
def preprocess_image(content):
    """Prend les bytes d'une image et retourne un array prétraité pour MobileNetV2"""
    try:
        # Ouvrir l'image depuis les bytes
        img = Image.open(io.BytesIO(content))
        
        # Redimensionner à 224x224 (taille attendue par MobileNetV2)
        img = img.resize((224, 224))
        
        # Convertir en array numpy
        img_array = np.array(img)
        
        # Si l'image est en niveaux de gris, convertir en RGB
        if len(img_array.shape) == 2:
            img_array = np.stack([img_array] * 3, axis=-1)
        elif img_array.shape[2] == 4:  # RGBA -> RGB
            img_array = img_array[:, :, :3]
        elif img_array.shape[2] == 1:  # 1 canal -> RGB
            img_array = np.concatenate([img_array] * 3, axis=-1)
        
        # Normaliser pour MobileNetV2 (preprocess_input spécifique)
        img_array = tf.keras.applications.mobilenet_v2.preprocess_input(img_array)
        
        return img_array
    except Exception as e:
        print(f"Erreur de prétraitement: {e}")
        # Retourner un array vide en cas d'erreur
        return np.zeros((224, 224, 3))

# 2. Fonction pour traiter un batch d'images
def process_batch(content_series):
    """Prend une série pandas de bytes et retourne les arrays prétraités"""
    # Convertir la série en liste
    content_list = content_series.tolist()
    
    # Prétraiter chaque image
    processed_images = [preprocess_image(content) for content in content_list]
    
    # Convertir en array numpy (batch_size, 224, 224, 3)
    batch_array = np.stack(processed_images)
    
    return batch_array

# 3. Test de la fonction de prétraitement sur une image
print("\n🧪 Test de prétraitement sur une image...")
test_image_row = images_with_labels.first()
test_content = test_image_row['content']

processed_test = preprocess_image(test_content)
print(f"✅ Test réussi !")
print(f"   • Taille originale : {len(test_content)} bytes")
print(f"   • Shape après prétraitement : {processed_test.shape}")
print(f"   • Type : {processed_test.dtype}")
print(f"   • Valeurs min/max : {processed_test.min():.2f}/{processed_test.max():.2f}")

# 4. Préparation de la fonction de création de modèle (à exécuter sur chaque worker)
def create_feature_extractor():
    """Crée l'extracteur de features MobileNetV2 sur chaque worker"""
    import tensorflow as tf
    from tensorflow.keras.applications import MobileNetV2
    from tensorflow.keras import Model
    
    # Charger le modèle
    base_model = MobileNetV2(
        weights='imagenet',
        include_top=True,
        input_shape=(224, 224, 3)
    )
    
    # Créer l'extracteur (sans dernière couche)
    feature_model = Model(
        inputs=base_model.input,
        outputs=base_model.layers[-2].output
    )
    
    # Geler les poids
    for layer in feature_model.layers:
        layer.trainable = False
    
    return feature_model

print("\n✅ Fonctions de prétraitement prêtes !")
print("   • preprocess_image() : prétraite une image")
print("   • process_batch() : prétraite un batch d'images")
print("   • create_feature_extractor() : crée le modèle sur chaque worker")
````

### ÉTAPE SUIVANTE : Définition de la Pandas UDF pour l'extraction de features distribuée.

#### Exécutin de cette cellule suivante ci-dessous pour définir et tester la UDF.

````python
# ÉTAPE 5 BIS - Approche simplifiée avec SCALAR au lieu de SCALAR_ITER

from pyspark.sql.functions import pandas_udf, PandasUDFType
from pyspark.sql.types import ArrayType, FloatType
import pandas as pd
import numpy as np

print("🔄 Approche simplifiée avec SCALAR UDF...")

# 1. Définir le type de retour
return_type = ArrayType(FloatType())

# 2. UDF simplifiée - SCALAR au lieu de SCALAR_ITER
@pandas_udf(return_type, PandasUDFType.SCALAR)
def extract_features_simple_udf(content_series):
    """
    UDF simplifiée pour extraction de features.
    SCALAR : traite une colonne entière en une fois (moins optimisé mais plus stable).
    """
    # Liste pour stocker les résultats
    all_features = []
    
    # Créer le modèle (une fois par partition)
    model = create_feature_extractor()
    
    # Traiter par petits batches pour éviter les problèmes mémoire
    batch_size = 10
    for i in range(0, len(content_series), batch_size):
        batch = content_series.iloc[i:i + batch_size]
        
        # Prétraiter le batch
        batch_images = []
        for content in batch:
            if content is not None:
                img = preprocess_image(content)
                batch_images.append(img)
            else:
                # Image vide si None
                batch_images.append(np.zeros((224, 224, 3)))
        
        if batch_images:
            # Convertir en array
            batch_array = np.stack(batch_images)
            
            # Extraire les features
            try:
                features = model.predict(batch_array, verbose=0, batch_size=batch_size)
                # Aplatir
                flattened = features.reshape(features.shape[0], -1)
                
                # Ajouter à la liste
                for row in flattened:
                    all_features.append(row.tolist())
            except Exception as e:
                print(f"Erreur prediction: {e}")
                # Ajouter des zéros en cas d'erreur
                for _ in range(len(batch_images)):
                    all_features.append([0.0] * 1280)
        else:
            # Batch vide
            for _ in range(batch_size):
                all_features.append([0.0] * 1280)
    
    # Retourner la série
    return pd.Series(all_features[:len(content_series)])

print("✅ UDF simplifiée définie !")

# 3. Test sur 2 images seulement
print("\n🧪 Test sur 2 images...")

test_df_small = images_with_labels.limit(2).cache()

# Appliquer l'UDF
test_result_small = test_df_small.select(
    "path",
    "label", 
    extract_features_simple_udf("content").alias("features")
)

print(f"✅ UDF appliquée sur {test_result_small.count()} images")

# Afficher le résultat sans collecter (éviter le crash)
print("\n📋 Aperçu du résultat (sans collect) :")
test_result_small.select("path", "label").show(truncate=False)

# Vérifier avec une action simple
print(f"\n🔍 Vérification avec count: {test_result_small.count()}")

print("\n🎯 UDF prête pour le traitement complet !")
````
